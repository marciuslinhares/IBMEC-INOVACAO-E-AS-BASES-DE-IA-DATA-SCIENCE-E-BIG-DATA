```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

# Configurar o Seaborn para criar gráficos mais bonitos
sns.set(style="whitegrid")
```

```python
# Carregar o dataset de clientes do shopping
url = 'https://raw.githubusercontent.com/tirthajyoti/Machine-Learning-with-Python/master/Datasets/Mall_Customers.csv'
data = pd.read_csv(url)

# Exibir as primeiras linhas do DataFrame
print(data.head())

```

```python

# Vamos utilizar as colunas Renda Anual (Annual Income) e Pontuação de Gasto (Spending Score) para aplicar o K-Means Clustering.


# Selecionar as colunas "Annual Income (k$)" e "Spending Score (1-100)"
X = data[['Annual Income (k$)', 'Spending Score (1-100)']]

# Exibir as primeiras linhas do DataFrame
print(X.head())
```

```python
# Inicializar o modelo K-Means com 5 clusters
kmeans = KMeans(n_clusters=5, random_state=42)

# Ajustar o modelo aos dados
kmeans.fit(X)

# Adicionar os rótulos dos clusters ao DataFrame
data['Cluster'] = kmeans.labels_

# Exibir as primeiras linhas com os clusters atribuídos
print(data.head())
```

```python
# Visualizar os clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(x=data['Annual Income (k$)'], y=data['Spending Score (1-100)'], hue=data['Cluster'], palette='viridis')
plt.title("Clusterização de Clientes do Shopping usando K-Means")
plt.xlabel("Renda Anual (Annual Income)($)")
plt.ylabel("Pontuação de Gasto (Spending Score) (1-100)")
plt.show()
```
